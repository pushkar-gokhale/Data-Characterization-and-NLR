```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Fit Equation
def fit(z,a,b,d):
    return a*z**2 - b*z + d
```

```python
# Calculating the R^2 value
from sklearn.metrics import r2_score
def calc(Rp,Rm):
    Rsqr = r2_score(Rp,Rm)
    #print()
    return Rsqr
```

```python
# Main Programme for Recursive Calculation
columns = list(mg_t.columns)
for i in columns[1:]:
    
    lc = np.log(mg_t[['Conductivity']]).values
    li = np.log(mg_t[i]).values
    lc_t = lc.T 
    
    c,cov = curve_fit(fit,li,lc_t[0])
    print(c)
    print()
    
    
    n = len(li)
    y = np.empty(n)
    for value in range(n):
        y[value] = fit(np.log(mg_t[i][value]),c[0],c[1],c[2])
    
    R = calc(y,lc)
    if R > 0.98:
        
        print('Fit Equation: (a*z**2 - b*z + d)')
        print()
        a = c[0]
        b = c[1]
        d = c[2]
        print('Final Parameters of Fit Equation')
        print('a: ',a)
        print('b: ',b)
        print('d: ',d)
        print("R^2: ",R)
        plt.plot(li,lc)
        plt.plot(li,y,'r')
        plt.title('Fit for '+str(i))
        plt.xlabel("log(Impedence)")
        plt.ylabel("log(conductivity)")
        break
    #print(mg_t[i])
```
