```python
import os, glob
import csv
```

```python
import pandas as pd
```

```python
path = "C:/Users/Shreeram/Desktop/Machine Learning/csv/Data/Sample Data Set #2"
```

```python
# Creating a path for all csv files
all_files = glob.glob(os.path.join(path, "*.csv"))
display(all_files)
```

```python
#Reading all .csv files
all_csv = (pd.read_csv(f, sep=',') for f in all_files)
```

```python
# Merging all .csv files
df_merged   = pd.concat(all_csv, ignore_index=True)
```

```python
mg = df_merged.drop(["#NAME?","|Z|/Ohm","Phase(Z)/deg","time/s","<Ewe>/V","<I>/mA","Cs/µF","Cp/µF"],axis = 1)
```

```python
mg = mg.drop(["cycle number","I Range","|Ewe|/V","|I|/A","THD Ewe/%","NSD Ewe/%","NSR Ewe/%","THD I/%","NSD I/%","NSR I/%","Re(Y)/Ohm-1","Im(Y)/Ohm-1","|Y|/Ohm-1","Phase(Y)/deg"],axis=1)
```

```python
mg = mg.groupby('freq/Hz')['Re(Z)/Ohm'].apply(lambda x: pd.Series(list(x))).unstack()
```

```python
mg.to_csv('data.csv')
```
