```python
import os, glob
import csv
```

```python
import pandas as pd
```

```python
path = "C:/Users/Shreeram/Desktop/Machine Learning/csv"
```

```python
# Creating a path for all csv files
all_files = glob.glob(os.path.join(path, "*.csv"))
display(all_files)
```

```python
#Reading all .csv files
all_csv = (pd.read_csv(f, sep=',') for f in all_files)
```

```python
# Merging all .csv files
df_merged   = pd.concat(all_csv, ignore_index=True)
df_merged
```

```python
# Grouping and Segrating y values with the corrosponding x
df_merged = df_merged.groupby('x')['y'].apply(lambda x: pd.Series(list(x))).unstack()

#df_merged.to_csv('merged.csv')
```

```python
# Saving it as new merged csv files
df_merged
df_merged.to_csv('merged.csv')
```

```python
# reading new merged csv file as merged
merged = pd.read_csv('merged.csv')
print(merged)
```

```python
# printing columns of merged csv file
print(merged.columns)
```

```python
# Renaming Columns
merged.rename(columns = {"x": "Input", 
                     "0":"Output1", "1":"Output2","2": "Output3","3": "Output4",
                        "4": "Output5","5": "Output6"}, 
                   inplace = True)
```
